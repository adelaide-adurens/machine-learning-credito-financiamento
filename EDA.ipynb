## Algumas análises iniciais das relações de algumas features com a inadimplência.

```python
'''
Verificação dos tipos de empréstimos
'''
df["NAME_CONTRACT_TYPE"].unique()
```

```python
'''
Percentual de empréstimos de cada tipo que tiveram problemas com o pagamento (1).
'''

for item in list(df["NAME_CONTRACT_TYPE"].unique()):
    print(f"Tipo de empréstimo: {item}\n")
    print(df[(df["NAME_CONTRACT_TYPE"] == str(item))]["TARGET"].value_counts(normalize=True))
    print("#"*80)
    print()
```

```python
'''
Verificação dos níveis educacionais
'''
df["NAME_EDUCATION_TYPE"].unique()
```

```python
'''
Percentual de empréstimos por nível educacional que tiveram problemas com o pagamento (1).
'''

for item in list(df["NAME_EDUCATION_TYPE"].unique()):
    print(f"Nível educacional: {item}\n")
    print(df[(df["NAME_EDUCATION_TYPE"] == str(item))]["TARGET"].value_counts(normalize=True))
    print("#"*80)
    print()
```

```python
'''
Análise das faixas de renda.
'''
df["AMT_INCOME_TOTAL"].describe()
```

```python
'''
Nota-se a presença de outliers, ou seja, algumas poucas rendas muito maiores que as demais.
'''
df["AMT_INCOME_TOTAL"].sort_values()
```

```python
'''
Divisão das rendas em "faixas de rendas". Valores determinados com base nas análises de mínimo, máximo e concentrações 
por quartis observados nas células anteriores.
'''
bins = [10000, 50000, 150000, 250000, 500000, 1000000, 117000000]
df["INCOME_RANGE"] = pd.cut(df["AMT_INCOME_TOTAL"], bins = bins)
```

```python
df["INCOME_RANGE"]
```

```python
'''
Percentual de empréstimos por faixa de renda que tiveram problemas com o pagamento (1).
'''

for item in list(df["INCOME_RANGE"].unique()):
    print(f"Faixa de renda: {item}\n")
    print(df[(df["INCOME_RANGE"] == item)]["TARGET"].value_counts(normalize=True))
    print("#"*80)
    print()
```

```python
'''
Verificação das ocupações
'''
df["OCCUPATION_TYPE"].unique()
```

```python
'''
Percentual de empréstimos por ocupação que tiveram problemas com o pagamento (1).
'''

for item in list(df["OCCUPATION_TYPE"].unique()):
    print(f"Ocupação: {item}\n")
    print(df[(df["OCCUPATION_TYPE"] == item)]["TARGET"].value_counts(normalize=True))
    print("#"*80)
    print()
```

```python
df["DAYS_BIRTH"]
```

```python
'''
Verificação das idades.
Observa-se que as idades estão em dias e negativas. Ao consultar o fórum da competição descobre-se que fizeram a conta
com base no momento da contratação. Então, para facilitar a visualização e entendimento, multipliquei por -1 e dividi por 
365 dias para obter a idade em anos.
'''

df['AGE_IN_YEARS'] = (df["DAYS_BIRTH"]* -1)/365
```

```python
df['AGE_IN_YEARS'].describe() 
```

```python
'''
Divisão das idades em "faixas de idade". Valores determinados com base nas análise da célula anterior.
'''

df["AGE_RANGE"] = df["AGE_IN_YEARS"].apply(lambda x: age_range(x))

df["AGE_RANGE"]
```

```python
'''
Percentual de empréstimos por faixa de idade que tiveram problemas com o pagamento (1).
'''

for item in list(df["AGE_RANGE"].unique()):
    print(f"Faixa de idade: {item}\n")
    print(df[(df["AGE_RANGE"] == item)]["TARGET"].value_counts(normalize=True))
    print("#"*80)
    print()
```

```python
'''
Visão consolidada da relação de algumas features analisadas até agora. Podemos ver o percentual de indivíduos com problemas
de pagamento do empréstimo relacionando a profissão, faixa salarial, faixa etária, nível educacional e tipo de empréstimo. Essa visão
também traz um perfil dos tomadores de crédito da base.
'''

show_all(df.groupby(["OCCUPATION_TYPE", "INCOME_RANGE", "AGE_RANGE", "NAME_EDUCATION_TYPE", "NAME_CONTRACT_TYPE"])
         [['TARGET']].agg(["count", "sum", "mean"]))
```
